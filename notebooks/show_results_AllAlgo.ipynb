```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time
import os
%matplotlib inline
from matplotlib.pylab import rcParams
```

```python
path = "dataset/results/allAlgo"
files = []
for file in os.listdir(path):
    if(".csv" in file and "result" in file):
        files.append(file)
#files = ["result0-buff10.csv","result1-buff10.csv"]
```

```python
measures = ["2019-05-13 00-00-01 scd30.csv","2019-05-28 01-00-01 scd30.csv","2019-06-13 01-00-02 scd30.csv",
          "2019-05-20 00-00-01 scd30.csv","2019-06-10 01-00-02 scd30.csv","2019-05-29 01-00-02 scd30.csv"]

dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m-%d %H-%M-%S')
data = []
for file in measures:
    data.append(pd.read_csv("dataset/results/ARpi1/"+file, parse_dates=['Time'], index_col='Time',date_parser=dateparse))
nbData = len(data)
for i in range(0,nbData):
    data[i] = data[i][data[i]["#CO2"]<=7000]
```

```python
index = ['Iteration','ARIMATime','ARIMAPerc','ARIMAConf','ARIMA_logTime','ARIMA_logPerc','ARIMA_logConf',
         'NewARIMATime','NewARIMAPerc','NewARIMAConf','NewARIMA_logTime','NewARIMA_logPerc','NewARIMA_logConf',
         'ARIMAUpTime','ARIMAUpPerc','ARIMAUpConf','ARIMAUp_logTime','ARIMAUp_logPerc','ARIMAUp_logConf',
         'ARTime','ARPerc','AR_logTime','AR_logPerc']
tmp = "Iteration,ARIMATime,ARIMAPerc,ARIMAConf,ARIMA_logTime,ARIMA_logPerc,ARIMA_logConf,NewARIMATime,NewARIMAPerc,NewARIMAConf,NewARIMA_logTime,NewARIMA_logPerc,NewARIMA_logConf,ARIMAUpTime,ARIMAUpPerc,ARIMAUpConf,ARIMAUp_logTime,ARIMAUp_logPerc,ARIMAUp_logConf,ARTime,ARPerc,AR_logTime,AR_logPerc"
```

```python
results = []
for i in range(0,len(files)):
    results.append(pd.read_csv(path+"/"+files[i], names = index))
    results[i] = results[i].set_index("Iteration")
print(results[0].head())
```

```python
for i in range(0,len(results)):
    print(measures[i])

    print("ARIMA")
    print("Time mean: "+str(results[i]["ARIMATime"].mean()))
    print("Percent mean: "+str(results[i]["ARIMAPerc"].mean()))
    print("\n")
    
    print("ARIMA Log")
    print("Time mean: "+str(results[i]["ARIMA_logTime"].mean()))
    print("Percent mean: "+str(results[i]["ARIMA_logPerc"].mean()))
    print("\n")
    
    print("New ARIMA")
    print("Time mean: "+str(results[i]["NewARIMATime"].mean()))
    print("Percent mean: "+str(results[i]["NewARIMAPerc"].mean()))
    print("\n")
    
    print("New ARIMA Log")
    print("Time mean: "+str(results[i]["NewARIMA_logTime"].mean()))
    print("Percent mean: "+str(results[i]["NewARIMA_logPerc"].mean()))
    print("\n")
    
    print("ARIMA Update")
    print("Time mean: "+str(results[i]["ARIMAUpTime"].mean()))
    print("Percent mean: "+str(results[i]["ARIMAUpPerc"].mean()))
    print("\n")
    
    print("ARIMA Update Log")
    print("Time mean: "+str(results[i]["ARIMAUp_logTime"].mean()))
    print("Percent mean: "+str(results[i]["ARIMAUp_logPerc"].mean()))
    print("\n")

    print("AR")
    print("Time mean: "+str(results[i]["ARTime"].mean()))
    print("Percent mean: "+str(results[i]["ARPerc"].mean()))
    print("\n")
    
    print("AR Log")
    print("Time mean: "+str(results[i]["AR_logTime"].mean()))
    print("Percent mean: "+str(results[i]["AR_logPerc"].mean()))
    print("\n")
    print("\n")
```

```python
rcParams['figure.figsize'] = 15, 6
xlabels = ["ARIMA", "ARIMA log", "New ARIMA", "New ARIMA Log", "ARIMA Update", "ARIMA Update Log", "AR", "AR log"]
z = 0
for res in results:
    #plt.figure()
    fig, ax = plt.subplots(2, sharex=True, gridspec_kw={'hspace': 0})
    z +=1
    fig.suptitle("Day "+str(z))
    dataTime = []
    dataTime.append(res["ARIMATime"])
    dataTime.append(res["ARIMA_logTime"])
    dataTime.append(res["NewARIMATime"])
    dataTime.append(res["NewARIMA_logTime"])
    dataTime.append(res["ARIMAUpTime"])
    dataTime.append(res["ARIMAUp_logTime"])
    dataTime.append(res["ARTime"])
    dataTime.append(res["AR_logTime"])
    ax[0].boxplot(dataTime)
    ax[0].set_ylabel("Seconds")
    #ax[0].set_xticklabels(xlabels, rotation=45, fontsize=8)
    
    dataPerc = []
    dataPerc.append(res["ARIMAPerc"])
    dataPerc.append(res["ARIMA_logPerc"])
    dataPerc.append(res["NewARIMAPerc"])
    dataPerc.append(res["NewARIMA_logPerc"])
    dataPerc.append(res["ARIMAUpPerc"])
    dataPerc.append(res["ARIMAUp_logPerc"])
    dataPerc.append(res["ARPerc"])
    dataPerc.append(res["AR_logPerc"])    
    ax[1].boxplot(dataPerc)
    ax[1].set_ylabel("Percent")
    ax[1].set_xticklabels(xlabels, rotation=45, fontsize=12)
```
